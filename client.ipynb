ICI pour compter et supprimer les clients ayant un bts qui n'appartient pas a la bd

```python
import pandas as pd


clients_df = pd.read_csv("client.csv")
bts_df = pd.read_csv("bts_final2.csv")

print(clients_df.columns)
print(bts_df.columns)
```

```python
bts_codes = bts_df['BTS'].unique()


clients_invalid_bts = clients_df[~clients_df['BTS'].isin(bts_codes)]


print(f"Nombre de clients avec un BTS inexistant : {len(clients_invalid_bts)}")

clients_invalid_bts.head()
```

```python
clients_cleaned = clients_df[clients_df['BTS'].isin(bts_codes)]


print(f"Nombre de clients aprÃ¨s nettoyage : {len(clients_cleaned)}")
```

```python
clients_cleaned.to_csv("clients_final.csv", index=False)
```
