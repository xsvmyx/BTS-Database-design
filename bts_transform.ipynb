```python
import pandas as pd
import json

# Charger les communes depuis le JSON
with open("communes.json", "r", encoding="utf-8") as f:
    communes = json.load(f)

# Indexer les communes par (name, wilaya_id)
commune_lookup = {
    (commune["name"].strip().upper(), int(commune["wilaya_id"])): commune["id"]
    for commune in communes
}

# Charger le CSV
df = pd.read_csv("bts_donnees_cleaned_ids.csv")

# Trouver le commune_id pour chaque ligne
commune_ids = []
for _, row in df.iterrows():
    key = (row["COMMUNE"].strip().upper(), int(row["WILAYA"]))
    commune_id = commune_lookup.get(key)
    if commune_id is None:
        print(f"❌ Commune non trouvée : {row['COMMUNE']} (wilaya {row['WILAYA']})")
    commune_ids.append(commune_id)

# Ajouter la nouvelle colonne et supprimer les anciennes
df["commune_id"] = commune_ids
df = df.drop(columns=["COMMUNE", "WILAYA"])

# Sauvegarder le nouveau CSV

df.to_csv("bts_with_commune_id.csv", index=False)
```

```python
import pandas as pd

# Lire le fichier CSV
df = pd.read_csv("bts_final.csv")

# Générer les lignes formatées BTS(...)
with open("formatted_bts.py", "w", encoding="utf-8") as f:
    for _, row in df.iterrows():
        code = row["BTS"]
        groupe = row["GROUPE"]
        latitude = row["Y"]
        longitude = row["X"]
        commune_id = int(row["commune_id"])  # convertir en int pour retirer le ".0"

        f.write(f'BTS(code="{code}", groupe="{groupe}", latitude={latitude}, longitude={longitude}, commune_id={commune_id}),\n')
```
